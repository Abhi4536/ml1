1. What is a parameter?
A parameter is an internal variable of a model that is learned from the training data. For example, in linear regression, the slope and intercept are parameters.

2. What is correlation?
Correlation measures the relationship between two variables—how one variable changes with respect to another. It ranges from -1 to +1.

3. What does negative correlation mean?
A negative correlation means that as one variable increases, the other decreases. Example: The more time you spend watching TV, the less time you may spend studying.

5. How does loss value help in determining whether the model is good or not?
The loss value tells us how far off our model’s predictions are from the actual values. A low loss means the model is performing well; a high loss indicates poor performance.

6. What are continuous and categorical variables?
Continuous variables: Numeric values that can take any value (e.g., height, weight, salary).

Categorical variables: Variables with fixed categories (e.g., gender, color, city).

10. What is a Test set?
The test set is a portion of data not seen by the model during training. It helps evaluate the generalization performance of the model.

17. What is causation? Difference between correlation and causation with an example?
Causation means one variable directly affects another.

Correlation is just a relationship, not a cause.

Example:

Correlation: Ice cream sales and drowning incidents increase in summer.

Causation: Summer causes both, but ice cream doesn’t cause drowning.
